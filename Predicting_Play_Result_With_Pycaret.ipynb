```python
import pandas as pd
df = pd.read_csv('drive/MyDrive/Colab Notebooks/plays.csv')
```

```python
df
```

```python
df.info()
```

```python
df.drop(df.columns[[0,1,2,3,4,]], axis=1, inplace=True)
```

```python
df['gameClock'] = df['gameClock'].apply(lambda x: round(int(x.split(':')[0]) + int(x.split(':')[1])/60, 2))
```

```python
df.info()
```

```python
from pycaret.regression import *

s = setup(df, target = 'playResult', normalize = True, session_id = 42)
```

```python
best = compare_models()
```

```python
finalize_model(best)
```

```python
evaluate_model(best)
```

```python
predict_model(best)
```
